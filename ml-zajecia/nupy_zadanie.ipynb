2.1 Zdefiniuj tablicę NumPy z sześcioma pomiarami tętna z opaski klienta po treningu, np.:
 65, 72, 80, 78, 90, 85
 Napisz program, który obliczy średnie tętno, minimalne oraz maksymalne tętno, wydrukuje w konsoli tylko te wartości tętna, które są wyższe od średniej oraz przygotuje nową tablicę, w której każde tętno jest zmniejszone o 5 (symulacja poprawy kondycji po kilku tygodniach treningu).

```python
import numpy as np

array = np.array([65, 72, 80, 78, 90, 85])
mean = array.mean()
mean
```

```python
print(array.max())
print(array.min())
```

```python
array[array > mean]
```
