2.2 Wykorzystując bibliotekę Pandas, wczytaj plik orders_raw.csv z danymi zamówień małego sklepu internetowego. Na podstawie wczytanego DataFrame wyświetl podstawowe informacje o danych (m.in. kilka pierwszych wierszy, liczbę wierszy i kolumn, typy danych) oraz policz, w których kolumnach występują braki danych. 
Następnie oczyść zbiór: usuń zbiorczo wszystkie pozycje w których występują brakujące wartości, usuń wiersze z oczywiście niepoprawnymi wartościami (np. Quantity ≤ 0 lub Price ≤ 0) oraz pozbądź się duplikatów. Dodaj nową kolumnę Total, która będzie zawierać łączną wartość zamówienia (ilość × cena), odrzuć tylko zamówienia, które nie, zostały wysłane (Shipped == "No"), a następnie sprawdź, z jakich krajów pochodzi najwięcej zamówień oraz który kraj generuje największą sumaryczną sprzedaż (możesz użyć grupowania i sortowania). Na końcu zapisz wyczyszczony i przefiltrowany zbiór do nowego pliku orders_clean.csv (bez kolumny indeksu).

```python
import pandas as pd

df = pd.read_csv('orders_raw.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.shape
```

```python
df = df[df['Price'] > 0]
df = df[df['Quantity'] > 0]
df['Price']
```

```python
df['Total'] = df['Price'] * df['Quantity']
df
```

```python
df = df[df['Shipped'] != 'No']
df
```

```python
df['Country'].value_counts()
```

```python
df.groupby('Country')['Quantity'].sum().sort_values(ascending=False)
```

```python
df.groupby('Country')['Total'].sum().sort_values(ascending=False)
```

```python
df.to_csv('orders_clean.csv',index=False)
```
