```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics, datasets, linear_model, model_selection

df = pd.read_csv('salary_Data.csv')
df
```

```python
df.describe()
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.draw()
```

```python
df['YearsExperience'].corr(df['Salary'])
```

```python
df['Salary'].corr(df['YearsExperience'])
```

```python
X = df['YearsExperience'].values
X = X.reshape(-1,1) #ilość rzedow i ilosc kolumn -> -1 oznacza ustaw sobie sam
print(type(X))
print(X.shape)
y = df['Salary']
```

```python
X_train, X_test, y_train, y_test =model_selection.train_test_split(X,y,test_size=0.2)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
predictions = model.predict(X_test)
predictions
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,predictions)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.draw()
```

```python
b = model.intercept_ #wysoksc
a = model.coef_[0] #slope

print(b) # dla osoby z 0 lat doświadczenia = 26000 (poczatek osi X)
print(a) # za każdy rok doświadczenia przyznaje 9199zł premii (nachylenie)
```

```python
import numpy as np
rmse = np.sqrt(metrics.mean_squared_error(y_test,predictions))
print(f'RMSE = {rmse} ({rmse/np.mean(predictions)*100:3.3}%)')
```

```python
pre_train_r2 = model.score(X_train, y_train)
pre_test_r2 = model.score(X_test, y_test)
print('Model determination train: ', pre_train_r2 )
print('Model determination test: ', pre_test_r2 )
```
