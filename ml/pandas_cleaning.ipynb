```python
core_cols = [
    "SLEPTIM1",   # godziny snu
    "_BMI5",      # BMI * 100
    "GENHLTH",    # ogólna ocena zdrowia
    "MENTHLTH",   # dni złego zdrowia psychicznego
    "PHYSHLTH",   # dni złego zdrowia fizycznego
    "DIABETE3",   # cukrzyca (tak/nie/inna)
    "CVDINFR4",   # przebyty zawał
    "EXERANY2",   # aktywność fizyczna
    "SMOKE100",   # ≥100 wypalonych papierosów w życiu
    "_AGE80"      # wiek w przedziałach
]

small = df[core_cols].copy()
```

```python
for col in core_cols:
    print(df[col].value_counts().head(10))

```

```python
for col in core_cols:
    print(col)
    print(small[col].isna().sum())
```

```python
small.describe()
```

```python
import numpy as np

# small.replace([77,99],np.nan,inplace=True)


# for col in core_cols:
#     if col != "SLEPTIM1":
#         small.loc[(small[col]== 7) | (small[col]== 9),col] = np.nan

missing_codes = {
    "SLEPTIM1": [77, 99],
    "GENHLTH": [7, 9],
    "MENTHLTH": [77, 99],
    "PHYSHLTH": [77, 99],
    "DIABETE3": [7, 9],
    "CVDINFR4": [7, 9],
    "EXERANY2": [7, 9],
    "SMOKE100": [7, 9],
    "_BMI5": [9999]
}

for col, codes in missing_codes.items():
    small[col] = small[col].replace(codes, np.nan)

# a 88 → 0 specjalnie
small["MENTHLTH"] = small["MENTHLTH"].replace(88, 0)
small["PHYSHLTH"] = small["PHYSHLTH"].replace(88, 0)
```

```python
for col in core_cols:
    print(f"col {col} : {(small[col] == 7).sum()}")
```

```python
small.describe()
```

```python
small = small.dropna()
```

```python
small.shape
```

```python
#korekta bmi

small["_BMI5"] = small["_BMI5"] / 100
small["_BMI5"].head()
```

```python
small_final = small.sample(5000,random_state=42)
small_final =small_final.rename(columns={"_BMI5":"BMI"})
small_final
```
